```python
# 필수
y = heart['output']
X = heart[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg' ,'thalachh', 'exng', 'oldpeak', 'slp', 'caa', 'thall']]
```

```python
# 필수
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size = 0.8,
                                                    random_state = 150,
                                                    stratify=y)
```

```python
y_test.value_counts()
```

```python
# 필수
# 로지스틱 회귀
from sklearn.linear_model import LogisticRegression
logi_model = LogisticRegression()

# train model
logi_model.fit(X_train,y_train)

# predict
logi_y_pred = logi_model.predict(X_test)
logi_y_pred
```

```python
# 필수
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)

# train model
knn_model.fit(X_train,y_train)

# predict
knn_y_pred = knn_model.predict(X_test)
knn_y_pred
```

```python
# 필수
# 의사결정나무
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=0)

# train model
dt_model.fit(X_train,y_train)

# predict
dt_y_pred = dt_model.predict(X_test)
dt_y_pred
```

```python
# 필수
from sklearn.metrics import accuracy_score

print(f"로지스틱 회귀 정확도: {round(accuracy_score(y_test, logi_y_pred),2)}")
print(f"KNN 정확도: {round(accuracy_score(y_test, knn_y_pred),2)}")
print(f"의사결정나무 정확도: {round(accuracy_score(y_test, dt_y_pred),2)}")
```

```python
# 필수
# 표준화
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
# 필수
print(X_train_scaled)
```

```python
# 필수
# 로지스틱 회귀-표준화
from sklearn.linear_model import LogisticRegression
logi_model = LogisticRegression()

# train model
logi_model.fit(X_train_scaled, y_train)

# predict
logi_y_pred = logi_model.predict(X_test_scaled)
logi_y_pred
```

```python
# 필수
# KNN-표준화
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)

# train model
knn_model.fit(X_train_scaled, y_train)

# predict
knn_y_pred = knn_model.predict(X_test_scaled)
knn_y_pred
```

```python
# 필수
# 의사결정나무-표준화
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=0)

# train model
dt_model.fit(X_train_scaled, y_train)

# predict
dt_y_pred = dt_model.predict(X_test_scaled)
dt_y_pred
```

```python
# 필수
from sklearn.metrics import accuracy_score

print(f"로지스틱 회귀 정확도: {round(accuracy_score(y_test, logi_y_pred),2)}")
print(f"KNN 정확도: {round(accuracy_score(y_test, knn_y_pred),2)}")
print(f"의사결정나무 정확도: {round(accuracy_score(y_test, dt_y_pred),2)}")
```
